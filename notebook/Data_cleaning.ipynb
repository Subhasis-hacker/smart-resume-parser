```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('UpdatedResumeDataSet.csv')
df.head()
```

```python
df.Category.value_counts()
```

```python
max_size=df['Category'].value_counts().max()

print("Original dataframe: \n",df)

balanced_df = df.groupby('Category').apply(lambda x : x.sample(max_size,replace=True)).reset_index(drop=True)
```

```python
df=balanced_df.sample(frac=1).reset_index(drop=True)
```

```python
df['Category'].value_counts()
```

```python
import re
```

```python
def clean(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)  
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)

    return cleanText

df['Resume']=df['Resume'].apply(lambda x: clean(x))
    
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df['Category']=le.fit_transform(df['Category'])
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
ti=TfidfVectorizer(stop_words='english')

req_text = ti.fit_transform(df['Resume'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(req_text,df['Category'],test_size=0.2,random_state=42)

```

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

knn=SVC()
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)

print(accuracy_score(y_test,y_pred))
```
